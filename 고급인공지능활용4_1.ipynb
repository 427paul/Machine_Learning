#데이터 시각화(Data Visualization)
##- 데이터 분석 결과를 쉽게 이해할 수 있도록 시각적으로 표현하고 전달

#Matplotlib
##-매트랩(Matlab)기능을 파이썬에서 그래도 사용하도록 하는 시각화 모듈
##- 엑셀 : 정형화된 차트, 그래프 작성, 다양한 함수 지원
##- 매트랩 : 포장(Wrapping), 맷플롯립 지원

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Pyplot
##- 맷플롯립을 이용할 때 가장 기본이 되는 객체
##- 파이플롯(pyplot)위에 그림(figure)객체를 올림
##- 그 위에 그래프에 해당하는 축(axes)을 올림
##- 그림 위에 축을 여려 장 올리면 여러 개의 그래프 작성


```python
x = range(100)
y = range(100)
plt.plot(x,y)
```

```python
x2 = range(100)
y2 = [np.cos(value) for value in x2]
#cos(x2)값들을 y2에 리스트로 저장
y2
```

```python
x3 = range(100)
y3 = [np.sin(value) for value in x3]
plt.plot(x3,y3)
```

```python
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.show
```

```python
fig, ax = plt.subplots()
#fig : 그래프 설정
#ax : 각 그래프 설정
x = range(100)
y = [np.cos(value) for value in x]

ax.plot(x,y)
ax.set(title='cos graph', xlabel = 'X', ylabel = 'Y')
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

ax1 = fig.add_subplot(1,2,1)
ax1.plot(x,y,color=  'blue')
ax2 = fig.add_subplot(1,2,2)
ax2.plot(x2,y2,c = 'g')
```

```python
fig, ax = plt.subplots(nrows = 3, ncols = 4)
```

```python
print(type(ax))
```

```python
x = np.linspace(-1,1,100)
#-1부터 1까지 100개
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig,ax = plt.subplots(2,2)
ax[0,0].plot(x,y1)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)
plt.show
```

#지정된 행렬에서 축(axes)의 위치 배열
##- 행 지정, 열 지정, 세번째 숫자(축의 위치)

```python
ax1 = plt.subplot(321)
plt.plot(x,y1)
ax2 = plt.subplot(322)
plt.plot(x,y1)
ax3 = plt.subplot(312)
plt.plot(x,y1)
ax4 = plt.subplot(325)
plt.plot(x,y1)
ax5 = plt.subplot(326)
plt.plot(x,y1)
plt.show()
```
