#색상 : color / c 매개변수 -> 색상변경
##1. RGB값을 사용해서 #을 붙여 16진법으로 색상 표현
##2. b, g, r, c, m, y, k, w 등 약어 입력

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value+100 for value in x2]

plt.plot(x1,y1,color='#000000')
plt.plot(x2,y2,c = 'c')
```

```python
plt.plot(x1,y1,c='b', linestyle='--')#dashed
plt.plot(x2,y2,c='r', ls='dotted')

plt.title("Two Lines")
plt.show()
```

#범례
##- 축 객체마다 범례를 설정할 수 있음
##- legend 함수를 사용하여 생성함
##shadow 매개변수 : 범례에 그림자 효과 추가
##loc 매개변수 : 범례의 위치 지정
##- 값은 center, upper, right 등 총 11가지
##- best라고 지정하면 적절한 위치에 범례가 놓임

```python
plt.plot(x1,y1,color='b', linestyle='dashed',label='line1')#파랑색 dashed선 line1
plt.plot(x2,y2,c='r', ls='dotted',label='line2')#빨간색 dotted선 line2
plt.legend(shadow = True, fancybox=False, loc='upper right')#범례 그림자 오른쪽 위에
plt.title('$y = ax + b$')#제목 $기울기$
plt.xlabel('$x line$')#x축 $기울기$
plt.ylabel('y line')#y축
plt.show()
```

#산점도(Scatter Plot)
##- 데이터 분포를 2차원 평면에 표현함
##- 매개변수 c : 포인트 색상 지정
##- marker : 포인트 모양 지정
##- size : 포인트 크기 지정
##- alpha : 포인터 불투명도 지정

```python
data1 = np.random.rand(512,2)#512행 2열로 균일분포 표준정규분포 난수 생성
data2 = np.random.rand(512,2)
plt.scatter(data1[:,0], data1[:,1], color = 'blue', marker = 'x')#1행의 512개 수를 x값, 2행의 512개 수를 y값으로 좌표 설정
plt.scatter(data2[:,0], data2[:,1], color = 'red', marker = 'o')#color가 c로 줄이지 못하는 경우도 있으니 색 설정할 때는 color라고 치자
plt.show()
```

```python
data = [[5.,25.,50.,20.],[4.,23.,51.,17.],[6.,22.,52.,19.]]
x = np.arange(0,8,2)#0부터 8전까지 2단위로.. [0,2,4,6]
plt.bar(x,data[0],color = 'blue',width=0.5)#[5,25,50,20] 파랑색, 두께 0.5로
plt.bar(x+0.5, data[1], color = 'g', width = 0.5)#0.5만큼 오른쪽에 초록색으로
plt.bar(x+1, data[2],color='r',width=0.5)#1만큼 오른쪽에 빨간색으로(color을 c로 줄이면 오류뜬다)
plt.show()
```

```python
N = 1000
X = np.random.normal(size = N)#1000개 난수
plt.hist(X,bins = 100)#난수 1000개를 구간 100개로 나눠서 산점도 표현
plt.show
```

```python
data = np.random.randn(100,5)#난수 100개를 5번씩 뽑기
plt.boxplot(data)#상자수염그래프 -> 데이터의 분포를 확인할 수 있다
plt.show()
```

#상자그림(Boxplot)
##사분위수를 시각화하여 데이터의 분포와 밀집 정도를 표현
##점 : 이상치(이상한 값)
##수염의 끝 : 최솟값, 최댓값, 상자길이의 1.5배
##상자의 끝 : 상위 25%, 상위7 5%
##IQR : 상자의 길이(상위25% ~ 상위 75%)
##상자 속 선 : 중간값(평균값 아니다)


##데이터를 작은 데이터부터 큰 데이터까지 정렬
##Q1(25%)부터 Q3(75%)까지 박스 형태로 위치시킴
##IQR(InterQuatile Range) : Q1 ~ Q3
##하단값(최솟값) : Q1 - 1.5 * IQR
##상단값(최댓값) : Q3 + 1.5 * IQR
##이상치(Outlier) : 상단 값과 하단 값을 넘어가는 값듯
