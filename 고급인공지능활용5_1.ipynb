# 선형 회귀는 연속된 변수를 예측하는 최적의 직선을 찾는 알고리즘
##장점
##- 모델이 간단하기 때문에 구현과 해석이 쉬움
##- 같은 이유로 모델링하는 데 오랜 시간이 걸리지 않음

##단점
##- 최신 알고리즘에 비해 예측력이 떨어짐
##- 독립변수와 예측변수의 선형 관계를 전제로 하기 때문에, 이러한 전제에서 벗어나는 데이터에서는 좋은 예측을 보여주기 어려움

#선형 회귀
##- 독립변수와 종속변수 간의 선형 관계를 전제로 한 모델
##- 구현 및 이해가 용이한 장점이 있음
#Null
##- 값이 비어 있는 것을 뜻함
##- 널값, Null value, 결측치 등으로 불리며, N/A, NA, NaN 등 다양한 방식으로 표현됨
##- Null값은 비어 있어서 알 수 없는 값이지, 0이 아닌 점에 주의해야 함

#사분위수(Qunatile)
##- 데이터를 오름차순으로 정리했을 때 25%, 50%, 75% 위치에서 확인한 값
##- 예를 들어 100개의 값들이 있다고 하면 가장 낮은 숫자부터 하나씩 세어 25번째 데이터, 50번째 데이터, 75번째 데이터가 각각 25%, 50%, 75%에 해당함
##- Q1, Q2, Q3라고도 표현함

#오버피팅
##- 모델이 학습셋에 지나치게 잘 맞도록 학습되어서 새로운 데이터에 대한 예측력이 떨어지는 현상을 의미함

#오퍼피팅/언더피팅 문제
##- 훈련셋을 사용하여 지나치게 학습된 모델을 만들면, 시험셋에 대한 예측은 오히려 떨어지는 경향이 있음
##- 비단 선형 회귀에서만의 문제가 아니라, 모든 머신러닝 모델에서 발생하는 문제임
##- 높은 편향 -> 언더피팅(데이터의 일부만 학습)
### 충분히 학습되지 않아 훈련셋에 대해서도 좋은 예측력이 나오지 않는 모델
### 편향은 알고리즘이 적절치 못하거나 중요한 부분을 놓쳐서 예측력이 낮을 때 높아짐
##- 높은 분산 -> 오버피팅(모든 데이터를 학습)
### 오버피팅된 모델에서는 새로운 데이터를 예측할 때 큰 오차 유발
### 분산은 훈련셋의 여러 노이즈를 모델에 포함될 경우 높아짐

##- 훈련셋조차 제대로 예측하지 못하는 언더피팅은 반드시 피해야 함
##- 머신러닝 알고리즘이 복잡해지면서 오버피팅은 빈번하게 발생하는데, 완전한 해결은 어렵고 줄여나간다는 개념으로 접근함
