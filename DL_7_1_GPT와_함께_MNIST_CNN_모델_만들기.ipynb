# ChatGPT
'파이토치(PyTorch)로 MNIST 데이터셋을 사용하는 간단한 CNN 분류 모델을 만들고,

학습 및 검증 정확도를 출력하는 코드를 코랩에서 바로 실행할 수 있게 작성해줘. 구글 코랩에서 실행할거야.'

```python
# ================================
# 1. 라이브러리 import
# ================================
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# GPU 사용 여부 확인
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ================================
# 2. 데이터셋 로드
# ================================
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

train_dataset = datasets.MNIST(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

test_dataset = datasets.MNIST(
    root='./data',
    train=False,
    download=True,
    transform=transform
)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1000, shuffle=False)

# ================================
# 3. CNN 모델 정의
# ================================
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 7 * 7, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        x = self.conv(x)
        x = self.fc(x)
        return x

model = CNN().to(device)

# ================================
# 4. Loss / Optimizer
# ================================
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# ================================
# 5. 학습 함수
# ================================
def train(model, loader):
    model.train()

    correct = 0
    total = 0

    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    accuracy = 100 * correct / total
    return accuracy

# ================================
# 6. 검증 함수
# ================================
def evaluate(model, loader):
    model.eval()

    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    accuracy = 100 * correct / total
    return accuracy

# ================================
# 7. 학습 실행
# ================================
epochs = 5

for epoch in range(epochs):

    train_acc = train(model, train_loader)
    val_acc = evaluate(model, test_loader)

    print(f"Epoch {epoch+1}/{epochs}")
    print(f"Train Accuracy: {train_acc:.2f}%")
    print(f"Validation Accuracy: {val_acc:.2f}%")
    print("-" * 30)
```

# Gemini
'파이토치(PyTorch)로 MNIST 데이터셋을 사용하는 간단한 CNN 분류 모델을 만들고,

학습 및 검증 정확도를 출력하는 코드를 코랩에서 바로 실행할 수 있게 작성해줘. 구글 코랩에서 실행할거야.'

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# 1. 디바이스 설정 (GPU 사용 가능 여부 확인)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'사용 중인 디바이스: {device}')

# 2. 하이퍼파라미터 설정
batch_size = 64
learning_rate = 0.001
num_epochs = 5

# 3. 데이터 전처리 및 로더 (Data Loader) 준비
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,)) # MNIST 데이터의 평균과 표준편차
])

# 학습용 데이터셋 및 로더
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)

# 검증(테스트)용 데이터셋 및 로더
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
test_loader = DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)

# 4. CNN 모델 정의
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        # 첫 번째 합성곱 층: 1채널(흑백) 입력 -> 16채널 출력
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)

        # 두 번째 합성곱 층: 16채널 입력 -> 32채널 출력
        self.conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # 완전 연결 층 (Fully Connected Layer)
        # 28x28 이미지가 두 번의 풀링(1/2 씩)을 거쳐 7x7 크기가 됨
        self.fc1 = nn.Linear(32 * 7 * 7, 128)
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128, 10) # 0~9까지 10개의 클래스

    def forward(self, x):
        x = self.pool1(self.relu1(self.conv1(x)))
        x = self.pool2(self.relu2(self.conv2(x)))
        x = x.view(-1, 32 * 7 * 7) # Flatten (1차원으로 펼치기)
        x = self.relu3(self.fc1(x))
        x = self.fc2(x)
        return x

model = SimpleCNN().to(device)

# 5. 손실 함수와 옵티마이저 설정
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# 6. 모델 학습 및 검증 루프
print("모델 학습을 시작합니다...")

for epoch in range(num_epochs):
    # --- 학습 (Training) 단계 ---
    model.train() # 모델을 학습 모드로 설정
    train_loss = 0.0
    correct_train = 0
    total_train = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # 순전파 및 손실 계산
        outputs = model(images)
        loss = criterion(outputs, labels)

        # 역전파 및 가중치 업데이트
        optimizer.zero_grad() # 기울기 초기화
        loss.backward()
        optimizer.step()

        # 학습 정확도 계산을 위한 데이터 수집
        train_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()

    avg_train_loss = train_loss / len(train_loader)
    train_accuracy = 100 * correct_train / total_train

    # --- 검증 (Validation) 단계 ---
    model.eval() # 모델을 평가 모드로 설정 (Dropout, BatchNorm 등이 평가 모드로 작동)
    val_loss = 0.0
    correct_val = 0
    total_val = 0

    with torch.no_grad(): # 검증 시에는 기울기를 계산하지 않음 (메모리 절약 및 속도 향상)
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)

            val_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total_val += labels.size(0)
            correct_val += (predicted == labels).sum().item()

    avg_val_loss = val_loss / len(test_loader)
    val_accuracy = 100 * correct_val / total_val

    # 결과 출력
    print(f'Epoch [{epoch+1}/{num_epochs}]')
    print(f'  [Train] Loss: {avg_train_loss:.4f}, Accuracy: {train_accuracy:.2f}%')
    print(f'  [Valid] Loss: {avg_val_loss:.4f}, Accuracy: {val_accuracy:.2f}%\n')

print("학습이 완료되었습니다!")
```
