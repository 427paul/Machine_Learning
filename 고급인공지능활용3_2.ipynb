#그룹별 집계
##-데이터로부터 동일한 객체를 가진 데이터만 따로 뽑음
##-기술통계 데이터를 추출함
##-엑셀의 피봇테이블(pivot table)기능과 비슷함
##-A반 수학 점수의 원본 데이터를 가지고 있을 경우 해당 데이터에서 그룹별 집계 방법
##ex) 같은 성별 학생들의 평균점수 집계, 50점 이상 학생 수 집계

#groupby 명령어
##-분할(split) : 같은 종류의 데이터끼리 나누는 기능
##-적용(apply) : 데이터 블록마다 연산 적용
##-결합(combine) : 연산 함수가 적용된 각 블록들을 합침

```python
lpl_data = {'Team':['Riders','Riders','Devils','Devils','Kings','kings','Kings','Kings','Riders','Royals','Royals','Riders'],
            'Rank':[1,2,2,4,4,5,1,1,2,4,1,2],
            'Year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
            'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(lpl_data)
```

```python
df
```

```python
df.groupby("Team")["Points"].sum()
#각 Team 별 Points의 합
```

```python
multi_groupby = df.groupby(['Team','Year'])["Points"].sum()
#Team별 Year별 Points의 합
```

```python
multi_groupby
```

```python
multi_groupby.index
```

```python
multi_groupby['Devils':'Kings']
#Devils부터 Kings까지
```

```python
multi_groupby.unstack()
#데이터프레임 형태로 풀기(groupby로 묶은 것들 풀기)
```

```python
multi_groupby.swaplevel()
```

```python
multi_groupby.swaplevel().sort_index()
```

```python
multi_groupby
#level 0 : Team
#level 1 : Year
```

```python
multi_groupby.groupby('Year').sum()
```

```python
multi_groupby.groupby('Team').sum()
```

#그룹화된 상태(grouped)
#분할(split) :
##-분할까지만 이루어진 상태
#get_group 함수
##해당 키 값을 기준으로 분할된 데이터프레임 객체를 확인
#집계(aggregation) :
##-요약된 통계 정보를 추출
##- agg 함수는 min, 넘파이 mean 등 기존 함수 그대로 적용함

```python
grouped = df.groupby("Team")
grouped
```

```python
grouped.get_group("Riders")
```

```python
grouped.agg(min)
```

```python
grouped.agg(np.mean)
```

#변환(transform)
##해당 정보를 변환함
##-키 값별로 요약된 정보가 아닌 개별 데이터로 변환을 지원함
##-적용 시점에서는 그룹화된 상태의 값으로 적용함

```python
grouped.transform(max)
```

```python
score = lambda x : (x - x.mean()) / x.std()
grouped.transform(score)
#각 요소 정규화
```

```python
df.groupby("Team").filter(lambda x : len(x) >= 3)
#Team값을 x로 놓았을 때, 길이가 3보다 큰 것들 검색
```

```python
df.groupby("Team").filter(lambda x : x["Points"].max() > 800)
#각 Team별로 Points의 최대값이 800점이 넘는 것
```
