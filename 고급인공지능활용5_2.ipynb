```python
df = pd.read_csv('insurance.csv')
df
```

```python
df.rename(columns={'expenses': 'charges'}, inplace=True)
df.drop(columns = ['region'], inplace=True)
df.loc[df['smoker'] == 'yes', 'smoker'] = 1
df.loc[df['smoker'] == 'no', 'smoker'] = 0
df.loc[df['sex'] == 'male', 'sex'] = 1
df.loc[df['sex'] == 'female', 'sex'] = 0
df['smoker'] = df['smoker'].astype(int)
df['sex'] = df['sex'].astype(int)
df
```

```python
X = df[['age','sex','bmi','children','smoker']]
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#데이터 셋 준비 : 8:2 샘플링하기
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
#test_size = 0.2 : 전체 데이터의 20%를 테스트 데이터로 사용
#test_size = 0.3 : 전체 데이터의 30%를 테스트 데이터로 사용
#test_size : 작은 값 → 훈련 데이터가 많아지고 테스트 데이터가 적어짐 → 모델 학습은 잘 되지만 일반화 성능 평가가 정확하지 않을 수 있음
#test_size : 큰 값 → 훈련 데이터가 적어지고 테스트 데이터가 많아짐 → 모델 학습이 부족할 수 있지만 테스트 평가는 신뢰도가 올라감

#random_state = 100 : 항상 같은 방식으로 데이터를 섞음
#random_state = None : 실행할 때마다 다르게 데이터를 섞음
#다른 random_state 값을 주면 학습/테스트 데이터가 다르게 나뉘고, 그에 따라 모델 성능도 다르게 나올 수 있음
#고정된 값을 쓰면 실험 결과의 재현성이 좋아짐. 논문이나 보고서 쓸 때 중요!
```

```python
#선형회귀 모델 불러오기
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
#model : 선형회귀(LinearRegression()) 속성을 부여함. 모델 객체를 생성
```

```python
#모델 훈련
#모델명.fit(독립변수, 종속변수)
model.fit(X_train, y_train)
#이제 모델에 새로운 데이터를 입력했을 때 예측이 가능해진 상태
```

```python
#모델 학습
pred = model.predict(X_test)
```

```python
#학습 결과물(예측한 값)
pred
```

```python
#학습결과를 실제 데이터와 비교해보기
comparision = pd.DataFrame({'actual':y_test,'pred':pred})
comparision
#'actual' : 실제값
#'pred' : 예측값
```
