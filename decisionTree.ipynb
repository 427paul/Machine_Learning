#결정 트리(Decision Tree)
##- 특정 변수에 대한 특정 기준값으로 데이터를 계속 분류해가면서 유사한 그룹으로 묶어내어 예측값을 만드는 알고리즘

#사과, 귤 예측 모델
# 질문이 5 이상인가?
##- 질문 하나로 둘을 명확하게 구분할 수 있을 것
##['종류(종속변수)'] : [사과, 귤, 귤, 사과, 사과]
##['지름'] : [6, 4. 5, 4, 7, 7.5]

#사과, 복숭아 구분 모델
##- 지름과 무게 모두 중첩되는 부분이 있음
##- 손쉽게 둘을 구분하는 기준점을 찾기 어렵지만, 몇 번의 분류를 수행한다면 가능
##['지름(독립변수1)'] : [7.5, 5, 7, 6.5, 7, 6]
##['무게(독립변수2)'] : [320, 240, 300, 230, 250, 200]
##['종류(종속변수)'] : [사과, 사과, 사과, 복숭아, 복숭아, 복숭아]
## 지름 >= 7 -> yes : 무게>= 240, No : 무게 >= 300

#분류과정 포인트
##- 첫 번째 분루과정 : 사용할 변수 선정과 기준점을 정하는 것이 매우 중요함
##- 두 번째 분류과정 : 각 상황에서 최적의 변수와 기준점을 찾아내야 함

#분류 결정 트리
##- DecisionTreeClassifier는 각 노드의 순도가 가장 높은 방향으로 분류함
##노드 : 결정트리 다이어그램에서 보이는 각 박스
##순도 : 한 노드 안에 여러 종류가 아닌 한 종류의 목표값만 있는 상태에 대한 지표
##- 지니인덱스와 교차 엔트로피를 평가하여 분류함

#지니 인덱스
##- 각 노드에 대해서 지니  인덱스가 계산됨
##p : 노드 안에 특정 아이템의 비율
##예) 사과 2, 복숭아 2인 경우 : 1 - [0.5^2 + 0.5^2] = 0.5
##예) 사과 1, 복숭아 3인 경우 : 1 - [0.25^2 + 0.75^2] = 0.375
##예) 사과 0, 복숭아 4인 경우 : 1 - [0^2 + 1^2] = 0
##->노드가 한쪽 아이템으로 완전히 분류된 경우 지니 인덱스가 0이며, 순도가 높다고 평가함
##->지니 인덱스의 최댓값은 0.5, 최솟값은 0이 나올 수 있음
##->사이킷런의 결정 트리에서 분류를 위한 기본 지수가 지니 인덱스임

#교차 엔트로피
##예) 사과 2, 복숭아 2인 경우 : 1
##예) 사과 1, 복숭아 3인 경우 : 약 0.81
##예) 사과 0, 복숭아 4인 경우 : 0
##->교차 엔트로피의 최댓값은 1, 최솟값은 0이 나올 수 있음
##->교차 엔트로피를 지니 인덱스 대신 활용할 수 있음

#회귀 결정 트리
##- 회귀는 연속형 변수를 대상으로 MSE를 평가 기준으로 이용함
##- MSE를 구하며, 결정 트리 회귀는 가장 낮은 MSE값이 나오도록 노드를 분류해나감
##- MSE는 사이킷런의 결정 트리 모델에서 기본값으로 설정된 평가 기준임
##- 필요에 따라 매개변수를 이용하여 MSE 대신 MAE나 Poisson 등으로 설정할 수도 있음

#데이터
##x = [10,20,35,5,15]
##y = [98,0,6,44,88]

#첫 번째 노드 : x>15
##False -> x = [10,5,15], y = [98,44,88]
##True -> x = [20, 35], y = [0,6]

#True 노드에서 MSE 계산
##실제값(y값)과 예측값(True에서 y의 평균)의 차이에 대한 계산
##y의 평균 : 3
##평균과 각 값의 차이 : 0-3 = -3, 6-3=3
##위의 값에 대한 제곱 후 합을 구함 : -3^2+3^ㅑ2 = 18
##위의 값을 관측치 개수만큼 나눔 : 18/2=9

```python
#매개변수 튜닝을 통해 성능 높이기
#결정 트리는 여러번 나눌수록(트리의 깊이가 깊어질수록) 분류가 잘 된다 -> 너무 많이 나눠지면 오버피팅 가능성 있음
#최대한 깊은 값으로 결정 트리를 나눠보자
model = DecisionTreeClassifier()#매개변수 지정X, 트리를 가장 깊게 분류
model.fit(X_train, y_train)
train_pred = model.predict(X_train)#훈련 데이터로 예측하기
test_pred = model.predict(X_test)#테스트 데이터로 예측하기
```

```python
print('Train score: ', accuracy_score(y_train, train_pred))
print('Test score: ', accuracy_score(y_test, test_pred))
#훈련셋은 성능이 좋지만, 다른 데이터셋을 사용하면 성능이 떨어진다 -> 오버피팅이 되었다는 뜻
```

```python
model = DecisionTreeClassifier(max_depth=5)#매개변수 설정
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Train score: ', accuracy_score(y_train, train_pred))
print('Test score: ', accuracy_score(y_test, test_pred))#train 셋보다 test 셋에서 성능이 좋은 모델이 더 좋다
```

```python
model = DecisionTreeClassifier(max_depth=7)#매개변수 설정. 트리를 더 깊게 만듦
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Train score: ', accuracy_score(y_train, train_pred))
print('Test score: ', accuracy_score(y_test, test_pred))
#전반적으로 성능이 더 좋아짐
```
