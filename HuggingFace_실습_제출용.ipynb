```python
from diffusers import DiffusionPipeline
from IPython.display import display

# 모델 로드 (CUDA가 없다면 .to("cpu")로 변경)
pipe = DiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4").to("cuda")

# 프롬프트 예시
prompt = sample['text']
image = pipe(prompt).images[0]

# 이미지 출력
display(image)
```

## Local Inference on GPU
Model page: https://huggingface.co/WarriorMama777/OrangeMixs

⚠️ If the generated code snippets do not work, please open an issue on either the [model repo](https://huggingface.co/WarriorMama777/OrangeMixs)
			and/or on [huggingface.js](https://github.com/huggingface/huggingface.js/blob/main/packages/tasks/src/model-libraries-snippets.ts) 🙏

```python
from diffusers import StableDiffusionPipeline
import torch
from IPython.display import display

pipe = StableDiffusionPipeline.from_pretrained(
    "Lykon/dreamshaper-7",  # 잘 작동하는 모델
    torch_dtype=torch.float16
).to("cuda")

prompt = sample['text']
image = pipe(prompt).images[0]

# Colab에서 보기
display(image)

```
